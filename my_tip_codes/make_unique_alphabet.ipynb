```
import re

def txt_to_oneSentence( string ):
  pattern = re.compile(r'\s+')
  return re.sub(pattern, '', string)
```

```
def str_to_unique( string ):
  unique_alpha = ""
  for i in string:
    if i not in unique_alpha:
      unique_alpha = unique_alpha + i
  return unique_alpha
```

```
text = open('1.txt', 'r')
```

```
one_text = txt_to_oneSentence( text.read() )
unique_str = str_to_unique( one_text )
unique_str
```

```
rst = list( unique_str )

f = open("unique_4.txt", 'w')

j = 0
for i in rst:
  # f.write( i + "\n" ) 
  if j==10 :
    f.write( i + "\n" )
    j = 0
  else :
    f.write( i )
  j = j+1

f.close()
```
